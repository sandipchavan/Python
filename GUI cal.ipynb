```python
#!/usr/bin/python
from Tkinter import *
mwin=Tk()

vexp=''
def button1():
	global vexp
	print "button 1"
	vexp+='1'
	etv.set(vexp)
def button2():
	global vexp
	vexp+='2'
	etv.set(vexp)
def button3():
	global vexp
	vexp+='3'
	etv.set(vexp)
def button4():
	global vexp
	vexp+='4'
	etv.set(vexp)
def button5():
	global vexp
	vexp+='5'
	etv.set(vexp)
def button6():
	global vexp
	vexp+='6'
	etv.set(vexp)
def button7():
	global vexp
	vexp+='7'
	etv.set(vexp)
def button8():
	global vexp
	vexp+='8'
	etv.set(vexp)
def button9():
	global vexp
	vexp+='9'
	etv.set(vexp)
def button0():
	global vexp
	vexp+='0'
	etv.set(vexp)
def add():
	global vexp
	vexp+='+'
	etv.set(vexp)
def minus():
	global vexp
	vexp+='-'
	etv.set(vexp)
def mul():
	global vexp
	vexp+='*'
	etv.set(vexp)
def div():
	global vexp
	vexp+="/"
	etv.set(vexp)
def eql():
	global vexp
	etv.set(str(eval(vexp)))
	vexp=""
	
mwin.title("calculator")
l1=Label()
l1.config(text="Simple Calculator",fg='red',bg='white')
l1.config(font=("Times New Roman",19,"bold italic"))
etv=StringVar()
e1=Entry(mwin,textvariable=etv)
b1=Button(mwin,text="1",command=button1)
b2=Button(mwin,text="2",command=button2)
b3=Button(mwin,text='3',command=button3)
b4=Button(mwin,text="4",command=button4)
b5=Button(mwin,text="5",command=button5)
b6=Button(mwin,text='6',command=button6)
b7=Button(mwin,text="7",command=button7)
b8=Button(mwin,text="8",command=button8)
b9=Button(mwin,text='9',command=button9)
b0=Button(mwin,text='0',command=button0)
badd=Button(mwin,text="+",command=add)
bminus=Button(mwin,text="-",command=minus)
bmul=Button(mwin,text="*",command=mul)
bdiv=Button(mwin,text="/",command=div)
beql=Button(mwin,text="=",command=eql)
l1.grid(row=0,columnspan=4,sticky=EW)
e1.grid(row=1,columnspan=4,sticky=EW)
b1.grid(row=2,column=0)
b2.grid(row=2,column=1)
b3.grid(row=2,column=2)
badd.grid(row=2,column=3)
b4.grid(row=3,column=0)
b5.grid(row=3,column=1)
b6.grid(row=3,column=2)
bminus.grid(row=3,column=3)
b7.grid(row=4,column=0)
b8.grid(row=4,column=1)
b9.grid(row=4,column=2)
bmul.grid(row=4,column=3)
b0.grid(row=5,column=0)
bdiv.grid(row=5,column=3)
beql.grid(row=5,column=1,columnspan=2,sticky=W)
mwin.mainloop()
```
